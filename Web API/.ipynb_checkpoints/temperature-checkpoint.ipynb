```python
con = lite.connect('city_weather.db')
cur = con.cursor()
```

```python
with con:
    cur.execute('DROP TABLE IF EXISTS daily_temp')
```

```python
with con:
    cur.execute('CREATE TABLE daily_temp (day_of_reading INT, new_york REAL, london REAL, toronto REAL, vancouver REAL, san_francisco REAL, seattle REAL, taipei REAL, sydney REAL);')
```

```python
end_date = datetime.now()
with con:
    for i in range(30):
        query_date = end_date - timedelta(days = i)
        cur.execute('INSERT INTO daily_temp (day_of_reading) VALUES (?)', (int(query_date.strftime('%s')),))
```

```python
for k,v in cities.iteritems():
    
    for i in range(30):
        query_date = end_date - timedelta(days=i)
        r = requests.get(url + v + ',' + query_date.strftime('%Y-%m-%dT12:00:00'))
        
        with con:
            cur.execute('UPDATE daily_temp SET ' + k + ' = ' + str(r.json()['daily']['data'][0]['temperatureMax']) + ' WHERE day_of_reading = ' + query_date.strftime('%s'))
con.close()
```

```python
con = lite.connect('city_weather.db')

df = pd.read_sql_query("SELECT * FROM daily_temp ORDER BY day_of_reading", con, index_col = "day_of_reading")
```

```python
import collections
```

```python
for col in df.columns:
    print col
```

```python
daily_change = collections.defaultdict(int)
for col in df.columns:
    temp_vals = df[col].tolist()
    temp_change = 0
    for k,v in enumerate(temp_vals):
        if k < len(temp_vals) - 1:
            temp_change += abs(temp_vals[k] - temp_vals[k+1])
    daily_change[col] = temp_change
```

```python
daily_change
```

```python
def keywithmaxval(d):
    v = list(d.values())
    k = list(d.keys())
    
    return k[v.index(max(v))]

max_city = keywithmaxval(daily_change)
max_city
```

```python
for col in df.columns:
    df.hist(column = col)
```

```python
df_tc = pd.DataFrame()
for col in df.columns:
    temp_vals = df[col].tolist()
    temp_delta_list = []
    temp_change = 0
    for k,v in enumerate(temp_vals):
        if k < len(temp_vals) - 1:
            temp_change = abs(temp_vals[k] - temp_vals[k+1])
            temp_delta_list.append(temp_change)
    df_tc[col] = temp_delta_list
```

```python
df_tc.plot(subplots=True, figsize=(8, 45)); plt.legend(loc='best')
```

```python
for col in df_tc:
    df_tc.hist(column = col)
```
