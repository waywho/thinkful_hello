```python
%matplotlib inline

import numpy as np
from pandas import Series, DataFrame
import pandas as pd
from datetime import datetime
```

```python
rng = pd.date_range('7/17/2012', periods = 100000, freq = 't')
rng
```

```python
ts = Series(np.random.randn(100000), index = rng)
ts
```

```python
ts.resample('30t', how = 'count', closed = 'left', label = 'left') #(freq, function, how)
```

```python
from datetime import timedelta
```

```python
resampled = ts.resample('30t', how = 'count',
                        loffset = timedelta(seconds = -1), closed = 'left', label = 'left') #(freq, function, how)
resampled
```

```python
resampled.index.hour
```

```python
ts[pd.date_range(ts.index[0], ts.index[-1], freq='30t')]
```

```python
ts.resample('M', how = 'median') #how can be sum, mean, max, or your own function
```

```python
timeit ts.resample('M', how = lambda x: x.mean())
```

```python
resampled = ts.resample('M', how = ['mean','median', 'count'])
resampled
```

```python
resampled['median']
```

```python
resampled = ts.resample('M', how = ['mean', np.max, 'count'])
resampled
```

```python
resampled = ts.resample('M', how = 'ohlc') #first, max, min, last
resampled
```

```python
resampled = ts.resample('5t', how = 'ohlc') #first, max, min, last
resampled=resampled[:5]
resampled
```

```python
hf = resampled.resample('t') #up sampling
hf.fillna(method='bfill', limit = 2) #forward filling ffill, or background filling bfill
```

```python
resampled = ts.resample('5t', how = ['first', 'max', 'min', 'last'])
resampled
```

```python
hf = resampled.resample('t')
limits = {'first': 1, 'max': 2, 'min': 3, 'last': 4}

hf.apply(lambda x: x.fillna(limit=limits[x.name]))
```

```python
resampled.resample('t', fill_method='ffill')
```

```python
stamp_naive = pd.Timestamp('2012-07-17 11:00')
stamp = pd.Timestamp('2012-07-17 11:00', tz = 'US/Central')
stamp.tz #pytz check timezone
```

```python
stamp2 = stamp_naive.tz_localize('US/Central')
```

```python
stamp3 = stamp2.tz_convert('Asia/Shanghai')
stamp3
```

```python
stamp3.value
```

```python
stamp2 == stamp3
```

```python
rng = pd.date_range('2012-03-01', periods=20, tz='US/Central')
ts = Series(range(20), index=rng)
ts
```

```python
rng = pd.date_range('2012-03-01', periods=20)
ts = Series(range(20), index=rng)
ts.tz_localize('utc').tz_convert('US/Central')
```

```python
rng = pd.date_range('2012-03-11', '2012-03-11 12:00', tz = 'utc', freq='h')
ts2 = Series(range(len(rng)), index = rng)
ts2.tz_convert('US/Central') #timezone and spring forward
```
